```python
import os
import pandas as pd
```

```python
df = pd.read_csv("ABIDEII-connectomes/ABIDEII_Composite_Phenotypic.csv", encoding="ISO-8859-1")
df
```

```python
list_of_subj = []

for root, dirs, files in os.walk("ABIDEII-connectomes"):
    for f in files:
        filepath = os.path.join(root, f)
        if filepath.find("AAL_space-") + 1:
            list_of_subj.append(int(filepath[filepath.index("sub-") + 4 : filepath.index("sub-") + 9]))

list_of_subj
```

```python
df = df.loc[df["SUB_ID"].isin(list_of_subj)]
df
```

```python
df = df.dropna(axis=1, how="all")
df
```

```python
df_dummy = df.dropna(axis=1)
df_dummy
```
